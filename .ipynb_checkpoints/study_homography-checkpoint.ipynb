```python
import cv2
import numpy as np
import datetime
```

```python
# 產生一個 1000x1000 白色影像 src
src = np.zeros((1001, 1001, 3), np.uint8)
src[:] = (255, 255, 255)
```

```python
import matplotlib.pyplot as plt
```

```python
# src影像畫上100x100尺寸的格子, 方便觀察
for i in range(0, 1001, 100):
    cv2.line(src, (0, i), (1000, i), (0, 0, 255), 3)
    cv2.line(src, (i, 0), (i, 1000), (0, 0, 255), 3)

    
# 順便打上4個紅色點
src = cv2.circle(src, (100,100), 10, (255, 0, 0), -1)
src = cv2.circle(src, (900,100), 10, (255, 0, 0), -1)
src = cv2.circle(src, (100,900), 10, (255, 0, 0), -1)
src = cv2.circle(src, (900,900), 10, (255, 0, 0), -1)
plt.plot(),plt.imshow(src),plt.title('src')  #顯示出來看看
plt.show()
cv2.imwrite('src_image.png', src)
```

```python
#rows,cols,ch = src.shape
```

```python
# 利用數個pairs of point, 找到 homography transformation matrix M
pts1 = np.float32([[100,100], [100,900], [900, 900], [900, 100]])
pts2 = np.float32([[1300,200], [400,1000], [1600, 1000], [1400, 200]])
```

```python
M, mask = cv2.findHomography(pts1, pts2)
M
```

```python
dst = cv2.warpPerspective(src, M, (2000, 2000))
cv2.imwrite('dst_image.png', dst)
```

```python
plt.plot(),plt.imshow(dst),plt.title('dst')  #顯示出來看看
plt.show()
```

```python
# 取得 invert homography transformation
inv_M = np.linalg.inv(M)
print M
```

```python
dst1 = dst.copy()
dst1 = cv2.circle(dst1, (1446, 1002), 30, (0, 255, 0), -1)
```

```python
# 將 dst 透過 inv_M 轉換成原圖src_prime
```

```python
src_prime = cv2.warpPerspective(dst, inv_M,(1000, 1000))
```

```python
plt.subplot(121), plt.imshow(dst), plt.title('dst')
plt.subplot(122), plt.imshow(src_prime), plt.title('src_prime')
plt.show()
```

```python
# 嘗試只轉換一個 point 的作法
def computeDstToSrc(x, y):
    result_pre = np.matrix(inv_M) * np.matrix([x, y, 1.0]).T
    result = result_pre / result_pre[2, 0]
    print("output src: ", result)
```

```python
# 嘗試只轉換一個 point 的作法
computeDstToSrc(1446, 1002)
```

```python
src_prime = cv2.circle(src_prime, (797, 900), 20, (0, 255, 0), -1)
```

```python
plt.subplot(121), plt.imshow(dst1), plt.title('dst1')
plt.subplot(122), plt.imshow(src_prime), plt.title('src_prime')
plt.show()
```
